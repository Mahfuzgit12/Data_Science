```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make sure to load the data before using it in the countplot
diabetes_data = pd.read_csv('diabetes.csv')  # Assuming 'diabetes.csv' is in the same directory

sns.countplot(x='Outcome', data=diabetes_data)
plt.show()
```

```python
diabetes_data.isnull().sum()
```

```python
coloumns_with_missing_values=["Glucose","BloodPressure","SkinThickness","Insulin","BMI"] # Correcting the typo in column names to "BloodPressure" and "Insulin"
missing_count = (diabetes_data[coloumns_with_missing_values] == 0).sum()
plt.figure(figsize=(10, 6))
sns.barplot(x=missing_count.index, y=missing_count.values)
plt.xlabel('Coloumns')
plt.ylabel('Count of Missing Values')
```

Unique Value Count:

```python
for column in diabetes_data:
  print(f"Unique values in {column}: {diabetes_data[column].nunique()}")
```

```python
#distribution of various parameters in the dataset except the target variable
cols=['Pregnancies','Glucose','BloodPressure','SkinThickness',
      'Insulin','BMI','DiabetesPedigreeFunction','Age']
# Replace 'df' with 'diabetes_data'
num=diabetes_data[cols]
for i in num.columns:
    plt.hist(num[i])
    plt.title(i)
    plt.show()
```
